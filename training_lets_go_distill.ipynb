```python
video_directory = '/content/gdrive/MyDrive/datasets/MSRVTT/videos/all'
def read_video_pyav(container, indices):
    '''
    Decode the video with PyAV decoder and frames.
    Args:
        container (`av.container.input.InputContainer`): PyAV container.
        indices (`List[int]`): List of frame indices to decode.
    Returns:
        result (np.ndarray): np array of decoded and frames of shape (num_frames, new_height, new_width, 3).
    '''
    frames = []
    container.seek(0)
    start_index = indices[0]
    end_index = indices[-1]


    for i, frame in enumerate(container.decode(video=0)):
        if i > end_index:
            break
        if i >= start_index and i in indices:
            frames.append(frame)

    return np.stack([x.to_ndarray(format="rgb24") for x in frames])


def sample_frame_indices(clip_len, frame_sample_rate, seg_len):
    '''
    Sample a given number of frame indices from the video.
    Args:
        clip_len (`int`): Total number of frames to sample.
        frame_sample_rate (`int`): Sample every n-th frame.
        seg_len (`int`): Maximum allowed index of sample's last frame.
    Returns:
        indices (`List[int]`): List of sampled frame indices
    '''
    converted_len = int(clip_len * frame_sample_rate)
    #print('con_len:', converted_len, 'seg_len:', seg_len)
    if converted_len >= seg_len:
      return [-1]
    end_idx = np.random.randint(converted_len, seg_len)
    start_idx = end_idx - converted_len
    indices = np.linspace(start_idx, end_idx, num=clip_len)
    indices = np.clip(indices, start_idx, end_idx - 1).astype(np.int64)
    return indices

class ImageCaptioningDataset(Dataset):
    def __init__(self, dataset, processor):
        self.dataset = dataset
        self.processor = processor

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        item = self.dataset[idx]
        # 
        # "datasets/MSRVTT/videos/all/video0.mp4"
        video_directory = '/content/gdrive/MyDrive/datasets/MSRVTT/videos/all'

        """train_list_path = "C:/Users/markk/Python_Projects/datasets/MSRVTT/videos/train_list_new.txt"
        video_files = [f for f in os.listdir(video_directory) if f.endswith(".mp4")]
        with open(train_list_path , 'r', encoding='utf-8') as file:
            # 
            lines = file.readlines()
            selected_line = lines[idx].strip()"""

        video_path = os.path.join(video_directory, f"{item['clip_name']}.mp4")
        #video_path = os.path.join(video_directory, f"{item['video_id']}.mp4")
        #print(idx)
        #print(item['video_id'])
        container = av.open(video_path)
        # sample frames
        num_frames = model.config.num_image_with_embedding

        if num_frames*4>=container.streams.video[0].frames:
          return -1

        indices = sample_frame_indices(
            clip_len=num_frames, frame_sample_rate=4, seg_len=container.streams.video[0].frames
        )
        if -1 in indices:
          return -1
                # print('num-frame: ', container.streams.video[0].frames)


        frames = read_video_pyav(container, indices)

        encoding = self.processor(images=list(frames), text=item["caption"], padding="max_length", return_tensors="pt")
        # remove batch dimension
        encoding = {k:v.squeeze() for k,v in encoding.items()}

        #print(encoding)

        return encoding
```

```python
def save_checkpoint(model, optimizer, scheduler, epoch, global_step, output_dir):
    checkpoint = {
        'epoch': epoch,
        'global_step': global_step,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'scheduler_state_dict': scheduler.state_dict()
    }
    torch.save(checkpoint, os.path.join(output_dir, f'checkpoint{epoch}.pth'))

def load_checkpoint(model, optimizer, scheduler, checkpoint_path):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    checkpoint = torch.load(checkpoint_path)
    model.load_state_dict(checkpoint['model_state_dict'])
    model.to(device) #problem
    optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    scheduler.load_state_dict(checkpoint['scheduler_state_dict'])
    epoch = checkpoint['epoch']
    global_step = checkpoint['global_step']
    return model, optimizer, scheduler, epoch, global_step
```

```python
# Create a new model with the modified configuration
from itertools import islice
model = AutoModelForCausalLM.from_config(existing_config)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
print(model.device.type)

#model = AutoModelForCausalLM.from_pretrained("C:/Users/markk/Python_Projects/Models", num_beams=4)
#print(model.config.num_hidden_layers)

processor = AutoProcessor.from_pretrained("microsoft/git-base-vatex")

train_dataset = ImageCaptioningDataset(dataset, processor)
val_dataset = ImageCaptioningDataset(dataset_val, processor)
batch_size = 1


train_dataloader = DataLoader(train_dataset, shuffle=False, batch_size=batch_size)
val_loader = DataLoader(val_dataset, shuffle=False, batch_size=batch_size)

optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)
optimizer.zero_grad()


epoch_num = 5
step_size = 15  # Adjust the learning rate every 15 steps
eval_size = step_size*150
gamma = 0.1     # Multiply the learning rate by gamma
warmup_ratio = 0.1

# Create a OneCycleLR scheduler
#scheduler = OneCycleLR(optimizer, max_lr=5e-3, total_steps=(epoch_num * int(len(train_dataloader)/step_size)) ,anneal_strategy='linear')
total_steps=epoch_num * int(len(train_dataloader)/step_size)
scheduler = get_scheduler(
                "linear",
                optimizer=optimizer,
                num_warmup_steps=ceil(total_steps*warmup_ratio),
                num_training_steps=total_steps,
            )


learning_rates = []
train_loss_list =[]
val_loss_list =[]
best_loss = 9999
total_loss = 0.0
total_val_loss = 0.0
step_idx = -1



batch = next(iter(train_dataloader))
for k,v in batch.items():
  print(k,v.shape)
prev_epoch=None
prev_step_idx=None
output_dir_current = "/content/gdrive/MyDrive/outputs/train_checkpoint/studentModel_current"
#####only for loading checkpoint###################################################################################################
#model, optimizer, scheduler, prev_epoch, prev_step_idx = load_checkpoint(model, optimizer, scheduler, output_dir_current+'/checkpoint0.pth')#
###################################################################################################################################
print(prev_epoch, prev_step_idx)
if prev_epoch is not None and prev_step_idx is not None:
  with open(f'/content/gdrive/MyDrive/outputs/loss.pkl', 'rb') as file:
    train_loss_list = pickle.load(file)
  with open(f'/content/gdrive/MyDrive/outputs/val_loss.pkl', 'rb') as file:
    val_loss_list = pickle.load(file)
  with open(f'/content/gdrive/MyDrive/outputs/lr.pkl', 'rb') as file:
    learning_rates = pickle.load(file)



###########training###############
model.train()

tensor_shape = (batch_size, 512)
input_ids_tensor= torch.full(tensor_shape, fill_value=101, dtype=torch.long).to(device)
mask = torch.zeros(512, device='cuda:0', dtype=torch.bool)
mask[0] = True
input_ids_tensor = input_ids_tensor * mask


kl_loss = nn.KLDivLoss(reduction="batchmean")

#teacher model logitsfile path
logits_directory = '/content/gdrive/MyDrive/datasets/MSRVTT/video_logits_batch/'

output_dir = f'/content/gdrive/MyDrive/outputs/Epoch_checkpoints'
output_dir_best = f"/content/gdrive/MyDrive/outputs/val_checkpoint/studentModel_best"
output_dir_current = f"/content/gdrive/MyDrive/outputs/train_checkpoint/studentModel_current"
#testmode = True

skipcomplete = 0

for epoch in range(epoch_num):
  print("Epoch:", epoch)
  if prev_epoch is not None and prev_epoch>epoch:
    continue

  for idx, batch in enumerate(tqdm(train_dataloader)):
    #print(idx)
    if batch==-1:
      print('skip')
      continue
    step_idx+=1

    if prev_step_idx is not None and prev_step_idx % len(train_dataloader)>=idx:
      continue
    if skipcomplete == 0: #0 means skip is not complete we need to tell it is complete now
      skipcomplete = 1
    if skipcomplete == 1:#if skip is complete we bench the step_idx to the proper step number
      if prev_step_idx is not None:
        step_idx += prev_step_idx * (prev_step_idx//len(train_dataloader))
      skipcomplete = 2 #we change the skipcomplete number to number other than 0 or 1, so that the next loop won't trigger the above if condition again

    labels = batch.pop("input_ids").to(device)
    pixel_values = batch.pop("pixel_values").to(device)
    # optimizer.zero_grad()
    outputs = model(input_ids=input_ids_tensor,
                    pixel_values=pixel_values,
                    labels=labels)

    item = dataset[idx]
    #print("idx: ", idx)
    #print("clip_name: ", item['clip_name'])
    video_filename = f"{item['clip_name']}.mp4.pkl"
    file_path = os.path.join(logits_directory, video_filename)
    if os.path.exists(file_path):
        with open(file_path, 'rb') as file:
            data = pickle.load(file)

    teacher_logits = torch.stack(data['logits'], dim=0)
    #print(len(teacher_logits))
    #Student Model Logits
    generated_ids = model.generate(pixel_values=pixel_values, max_length = 10, return_dict_in_generate=True, output_scores=True)
    student_logits = generated_ids.scores
    #print(len(generated_ids.scores))
    #print(len(generated_ids.scores[0]))
    alpha = 0.8
    temperature = 1.0
    distillation_loss = 0
    #print(generated_ids.sequences)

    for i in range(min(len(generated_ids.scores),len(teacher_logits))):
        for beam in range(existing_config.num_beams):
            q = F.log_softmax(student_logits[i][beam][:] / temperature, dim=-1).to(device)
            p = F.softmax(teacher_logits[i][beam][:], dim=-1).to(device)
            distillation_loss+= kl_loss(q,p)
    distillation_loss/= (min(len(generated_ids.scores),len(teacher_logits))*existing_config.num_beams)

    #print(pixel_values, labels)
    #print(f"distillation_loss: {distillation_loss}")
    #print(f"student_outputs.loss: {outputs.loss}")
    loss = alpha*distillation_loss + (1-alpha)*outputs.loss
    #loss = outputs.loss

    #print("Loss:", loss.item())
    total_loss+=loss

    if (step_idx%step_size==step_size-1):
      total_loss.backward()
      t = total_loss.detach().item()/step_size

      train_loss_list.append(t)
      total_loss = 0.0

      optimizer.step()
      # Step the learning rate scheduler
      learning_rates.append(scheduler.get_last_lr()[0])
      scheduler.step()

      optimizer.zero_grad()
    del labels
    del pixel_values
    del outputs
    # Record the learning rate for visualization

    ###########validation###############
    if(step_idx%eval_size==eval_size-1):
    #if(testmode):
      print(step_idx)
      model.eval()
      val_loss = 0.0
      #correct = 0
      #total = 0
      with torch.no_grad():
        print("Epoch:", epoch)
        for val_batch in tqdm(val_loader):
            if  val_batch == -1:
                continue
            input_ids = val_batch.pop("input_ids").to(device)
            pixel_values = val_batch.pop("pixel_values").to(device)
            outputs = model(input_ids=input_ids_tensor,
                        pixel_values=pixel_values,
                        labels=input_ids)

            val_loss += outputs.loss.detach().item()
            #print("Loss:", loss.item())
            del input_ids
            del pixel_values
            del outputs
      val_loss/=len(val_loader)
      if(val_loss < best_loss):
        best_loss = val_loss
        model.save_pretrained(output_dir_best)
        # save_checkpoint(model, optimizer, scheduler, epoch, step_idx, output_dir_best)
      #save current model
      # model.save_pretrained(f"/content/drive/MyDrive/outputs/train_checkpoint/studentModel_current")
      save_checkpoint(model, optimizer, scheduler, epoch, step_idx, output_dir_current)
      val_loss_list.append(val_loss)


      with open(f'/content/gdrive/MyDrive/outputs/loss.pkl', 'wb') as file:
          pickle.dump(train_loss_list, file)
      with open(f'/content/gdrive/MyDrive/outputs/val_loss.pkl', 'wb') as file:
          pickle.dump(val_loss_list, file)
      with open(f'/content/gdrive/MyDrive/outputs/lr.pkl', 'wb') as file:
          pickle.dump(learning_rates, file)
  # save checkpoint for each epoch, incase we want to resume training from that epoch
  # save_checkpoint(model, optimizer, scheduler, epoch, step_idx, output_dir)
  # model.save_pretrained(f"/content/drive/MyDrive/outputs/studentModel")

model.save_pretrained(f"/content/gdrive/MyDrive/outputs/studentModel")


with open(f'/content/gdrive/MyDrive/outputs/loss.pkl', 'wb') as file:
    pickle.dump(train_loss_list, file)
with open(f'/content/gdrive/MyDrive/outputs/val_loss.pkl', 'wb') as file:
    pickle.dump(val_loss_list, file)
with open(f'/content/gdrive/MyDrive/outputs/lr.pkl', 'wb') as file:
    pickle.dump(learning_rates, file)
```
